# Get video from environment

## Pong

```python
env = gym.make("Pong-v0")
env = Recorder(env, './video')

observation = env.reset()
terminal = False

while not terminal:
  action = env.action_space.sample()
  observation, reward, terminal, info = env.step(action)
env.play()
```
